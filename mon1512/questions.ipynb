Write a Python program that takes user name, age, and salary as input, converts them into appropriate datatypes, and prints the information using proper string formatting.


```python
name = input("Enter name: ")
age = int(input("Enter age: "))
salary = float(input("Enter salary: "))

print("Name: {}, Age: {}, Salary: {:.2f}".format(name, age, salary))
```

Write a program that takes marks of 5 subjects from the user, validates that each mark is between 0 and 100, and then calculates and prints the total and average marks.

```python
total = 0
count = 0

while count < 5:
    mark = int(input(f"Enter marks for subject {count + 1}: "))
    if 0 <= mark <= 100:
        total += mark
        count += 1
    else:
        print("Invalid marks. Enter between 0 and 100.")

average = total / 5
print("Total:", total)
print("Average:", average)
```

Write a program that takes 10 numbers as input using a loop and counts how many numbers are even and how many are odd. Display the result using formatted output.

```python
even_count = 0
odd_count = 0

for i in range(10):
    num = int(input("Enter number: "))
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
```

Write a program that stores a fixed number and repeatedly asks the user to guess the number using a while loop. The program should stop when the correct number is guessed and display the number of attempts.


```python
fixed_number = 7
attempts = 0

while True:
    guess = int(input("Guess the number: "))
    attempts += 1
    if guess == fixed_number:
        print("Correct! Attempts:", attempts)
        break
```

Write a program that displays a menu with options to square a number, cube a number, or exit. Use conditional statements to perform the selected operation and repeat the menu until exit is chosen.


```python
while True:
    print("1. Square")
    print("2. Cube")
    print("3. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        n = int(input("Enter number: "))
        print("Square:", n * n)
    elif choice == 2:
        n = int(input("Enter number: "))
        print("Cube:", n * n * n)
    elif choice == 3:
        break
    else:
        print("Invalid choice")
```

Write a program that takes a number as input and checks whether it is a prime number using loops. Use the break statement when a factor is found.

```python
num = int(input("Enter number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

Write a program that takes an integer n as input and prints a right-angled triangle star pattern of height n using nested loops.

```python
n = int(input("Enter height: "))

for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()
```

Write a program that loops from 1 to 30 and skips numbers divisible by 5 using the continue statement. Print all other numbers.

```python
for i in range(1, 31):
    if i % 5 == 0:
        continue
    print(i)
```

Write a program that takes a password as input and checks whether the password length is at least 8 characters. Display whether the password is valid or invalid.


```python
password = input("Enter password: ")

if len(password) >= 8:
    print("Valid password")
else:
    print("Invalid password")
```

Write a program that continuously takes numbers from the user, adds them to a total, and stops when the user enters 0. Finally, print the total sum.

```python
total = 0

while True:
    num = int(input("Enter number: "))
    if num == 0:
        break
    total += num

print("Total sum:", total)
```
