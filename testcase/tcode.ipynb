```python
def check_loan_eligibility(age,income,loan):
    if age <=0 or income <0 or loan<0:
        return "Invalid Input"
    elif age <21 or age >60:
        return "Not Eligible"
    elif income <20000:
        return "Not Eligible"
    elif loan > 0.4 * income:
        return "Not Eligible"
    else:
        return "Eligible"
    
```

```python
f = open('./bank_loan_eli/test_cases.txt', 'r')
data = f.read()
f = f.split('\n')

f = f[1:]

for row in f:
    age, income, loan = row.split(',')
    age=int(age)
    income=float(income)
    loan=float(loan)

    result = check_loan_eligibility(age,income,loan)
    print("Age: ",age)
    print("Income: ",income)
    print("Loan: ",loan)
    print("Result: ",result)
    print("---------------")
    
```
