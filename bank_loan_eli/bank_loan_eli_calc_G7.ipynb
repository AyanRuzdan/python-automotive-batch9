```python
def calculate_loan():
    loan_amount = float(input("Enter Loan Amount : "))
    yearly_interest = float(input("Enter Annual Interest in %: "))
    years = int(input("Enter Loan Tenure (Years): "))
    monthly_interest = yearly_interest / (12 * 100)    # Convert yearly interest to monthly interest
    months = years * 12                                                    # Convert years to months
    emi = (loan_amount * monthly_interest * (1 + monthly_interest) ** months) / ((1 + monthly_interest) ** months - 1)             #  formula for emi
    total_payment = emi * months
    total_interest = total_payment - loan_amount                        # Total payment and interest
    print("\n--- Loan Calculation Details ---")  
    print("Monthly EMI        :",emi,)
    print("Total Interest     : ",total_interest, )
    print("Total Amount Paid  : ",total_payment,)                               # Output
```

```python
def check_age(age):
    return age >= 18
```

```python
def check_income(income):
    return income >= 25000
```

```python
def check_existing_loans(existing_loans):
    return existing_loans <= 100000
```

```python
def check_loan_eligibility(age=18,mon_inc=25000,existing_loans=0):
    if check_age(age) == True:
        if check_income(mon_inc) == True:
            if check_existing_loans(existing_loans) == True:
                print("You are eligible for the loan")
            else:
                print("You already have a large loan")
        else:
            print("Your income is not eligible for the loan")
    else:
        print("Your age is not eligible for the loan")
```

```python
age = int(input("Enter age: "))
mon_inc = int(input("Enter monthly income: "))
existing_loans = int(input("Enter existing loan amount: "))
```

```python
print(check_loan_eligibility(age,mon_inc,existing_loans))
```
